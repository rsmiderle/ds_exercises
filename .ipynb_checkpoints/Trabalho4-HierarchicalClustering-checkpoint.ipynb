## 2. Gerar o dendograma com um ou mais métodos hierarquicos aglomerativos.

```
# Comparando os métodos no dendograma
# Methods: https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.linkage.html#scipy.cluster.hierarchy.linkage

plt.figure(figsize=(35,5));
# methods = ['single', 'ward', 'complete', 'average', 'weighted', 'centroid', 'median'] single dá erro "maximum recursion depth exceeded while calling a Python object"
methods = ['ward', 'complete', 'average', 'weighted', 'centroid', 'median']
_len = len(methods)
for j in range(_len):
  plt.subplot(1, _len, j + 1)
  plt.title(methods[j])
  dend = shc.dendrogram(shc.linkage(df_scaled, method=methods[j])) 
plt.show()
```

## 3. Gerar o agrupamento para pelo menos um algoritmo hierárquico aglomerativo.

### Ward

```
cluster = AgglomerativeClustering(n_clusters=8, affinity='euclidean', linkage='ward')
_hc = cluster.fit_predict(df_scaled)
```

```
groups = list()
values = list()
for i in range(8):
  groups.append(str(i))
  values.append(df_scaled[_hc==i, 0].sum())
```

```
plt.bar(groups, values)
plt.title('Ward')
plt.xlabel('Quantidade')
plt.ylabel('Grupos')
plt.show()
```

### Complete

```
cluster = AgglomerativeClustering(n_clusters=8, affinity='euclidean', linkage='complete')
_complete = cluster.fit_predict(df_scaled)
```

```
groups = list()
values = list()
for i in range(8):
  groups.append(str(i))
  values.append(df_scaled[_complete==i, 0].sum())
plt.bar(groups, values)
plt.title('Complete')
plt.xlabel('Quantidade')
plt.ylabel('Grupos')
plt.show()
```

## 4. Encontrar uma solução ótima de agrupamento a partir da visualização do dendograma e se necessário utilizar o SSE.
   8 grupos

```
scaler = StandardScaler()
df_kmeans = scaler.fit_transform(df_scaled)
```

```
def kmeans(numero_de_clusters, generos):
  modelo = KMeans(n_clusters=numero_de_clusters)
  modelo.fit(generos)
  return [numero_de_clusters, modelo.inertia_/100]
```

```
resultado=[]
for numero_de_grupos in range (1,21):
    resultado.append(kmeans(numero_de_grupos, df_kmeans))
```

```
resultado = pd.DataFrame(resultado, columns=['grupos', 'inertia'], index=list(range(1, 21)))
```

```
resultado.inertia.plot(x=resultado.grupos, y=resultado.inertia,figsize=(15, 5),xticks=resultado.grupos, xlabel="Clusters", ylabel="SSE")
```

## Diferenças K-means e Hierarchical clustering:
* Com Hierarchical clustering pode-se identificar mnais facilmente a presença dos outliers;
* Identificação do números de clusters nos pareceu mais clara com K-means;
* Hierarchical clustering permitiu uma melhor visualização gráfica por grupo;
* K-means gerou grupos com menos variações de quantidade.
