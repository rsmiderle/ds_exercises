### Resolvendo o balanceamento do Dataset

<p> Serão aplicadas técnicas undersample e oversample afim de melhorar o balanceamento do Dataset.
<p> Após cada ensaio será aplicado um modelo de regressão logística para avaliara a eficácia do Dataset e aí então decidir pela melhor técnica de balanceamento.

#### Aplicando modelo de regressão no dataset inteiro para ter uma métrica de quão eficiente nosso dataset é antes de aplicar qualquer técnica para melhorar o balanceamento

<p> Definiçaõ da função "testa_modelo", que avalia a eficácia do dataset.

```python
def testa_modelo(df):
    # separar variáveis entre X e y
    Xa = df.drop('STATUS_GOOD', axis=1)
    ya = df['STATUS_GOOD']

    # dividir o dataset entre treino e teste
    X_train, X_test, y_train, y_test = train_test_split(Xa, ya, stratify=ya, shuffle=True)
    # instanciar e treinar um modelo de Regressão Logística
    model = LogisticRegression()
    model.fit(X_train, y_train)

    # fazer as previsões em cima dos dados de teste
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)

    # plotar a matrix de confusão
    skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=True)

    # imprimir relatório de classificação
    print("Relatório de Classificação:\n", classification_report(y_test, y_pred, digits=4))

    # imprimir a acurácia do modelo
    print("Acurácia: {:.4f}\n".format(accuracy_score(y_test, y_pred)))
```

```python
testa_modelo(df2)
```

<p>Com o dataset completo a acurácia foi de 98,4%, porém, não obteve sucesso em prever quando o STATUS_GOOD será igual a 0 (mau pagador). Deu overfit para prever quando o valor será 1 e não conseguiu prever quando o valor será 0

# AGRUPANDO COM K-MEANS

```python
from sklearn.cluster import k_means
from sklearn.metrics import davies_bouldin_score as dbs
from sklearn.metrics.cluster import calinski_harabasz_score as chs
from sklearn.model_selection import cross_val_score as cvs

def makeKmeans(df_kmeans, n_clusters):
    # Executando o algoritmo k-means.
    centroides, rotulos_kmeans, sse = k_means(df_kmeans, 
                                              n_clusters=n_clusters, 
                                              init='random')
    return centroides, rotulos_kmeans, sse


def kmeansSSEEvaluating(df, numGrupos):
    sses = []
    for n_groups in range (2,numGrupos+1):
        sses.append(makeKmeans(df,n_groups)[2])
    plt.plot(sses)
    

def kmeansDBSEvaluating(df, numGroups):
    dbss = []
    for n_groups in range (2,numGroups+1):
        
        c1, r1, s1 = makeKmeans(df,n_groups)
        score = dbs(df,r1)
        dbss.append(score)
    plt.plot(dbss)
    

def kmeansCHBEvaluating(df, numGroups):
    chss = []
    for n_groups in range (2,numGroups+1):
        
        c1, r1, s1 = makeKmeans(df,n_groups)
        score = chs(df,r1)
        chss.append(score)
    plt.plot(chss)
```

```python
kmeansSSEEvaluating(df_scaled, 100)
```

```python
kmeansDBSEvaluating(df_scaled,100)
```

# Agrupando usando Algoritmo Hierarquico

```python
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import fcluster

def makeDfDendo(data):
  dataset = pd.DataFrame({'AMT_INCOME_TOTAL': data['AMT_INCOME_TOTAL'], 'DAYS_BIRTH': data['DAYS_BIRTH'], 'DAYS_EMPLOYED': data['DAYS_EMPLOYED']})
  return dataset

def makeHierarchy(df, method):
    h = shc.linkage(df, method=method)
    return h

def makeDendogram(h, title):
    plt.figure(figsize=(10, 7))  
    plt.title(title)  
    dend = shc.dendrogram(h)
    
def makeAgglomerativeClusteringScatter(n_clusters, affinity, linkage, df):
    cluster = AgglomerativeClustering(n_clusters=n_clusters, affinity=affinity, linkage=linkage)
    cluster.fit_predict(df)  
    plt.figure(figsize=(10, 7))  
    plt.scatter(df[df.columns[0]], df[df.columns[1]], c=cluster.labels_, cmap='rainbow')  
    plt.xlabel(df.columns[0], fontsize=12)
    plt.ylabel(df.columns[1], fontsize=12)
```

```python
# Colunas selecionadas AMT_INCOME_TOTAL, DAYS_BIRTH, DAYS_EMPLOYED
data_under = df2.iloc[:, [1,2,3]]  

df_dendo_under = makeDfDendo(pd.DataFrame(data_under))
h_under = makeHierarchy(df_dendo_under, 'ward')
makeDendogram(h_under, 'Dendograma')
```

```python
makeAgglomerativeClusteringScatter(n_clusters=5,affinity='euclidean', df=df_dendo_under, linkage='ward')
```
