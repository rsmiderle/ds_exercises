```python
%config Completer.use_jedi = False
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('application_record.csv')
```

<h1> 3. Escolher um atributo categóricos e utilizar a função OneHotEncoder  (Links para um site externo.) para transformar em codificação inteira-binária os valores nominais

```python
from sklearn.utils import resample
data_df_res = df
resample(df,data_df_res, n_samples=50, replace=False)
data_df_res.head(3)
```

<h3>Identificar as features categóricas </h3>

```python
data_df_res.nunique()
```

<b>Fazendo o One Hot Encoder para as features NAME_INCOME_TYPE e NAME_EDUCATION_TYPE. Cada uma delas tem 5 valores possíveis, devendo resultar em 10 novas colunas binárias

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(data_df_res[['NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE']]).toarray())
new_df = data_df_res.join(enc_df)
new_df
```
